Preparando datos:

Lo primero que hacemos con un modelo es cargar datos.

```python
import pandas as pd

# Hacer un diccionario de datos para tamaños de botas
# y tamaño del arnés en cm
data = {
    'boot_size' : [ 39, 38, 37, 39, 38, 35, 37, 36, 35, 40, 
                    40, 36, 38, 39, 42, 42, 36, 36, 35, 41, 
                    42, 38, 37, 35, 40, 36, 35, 39, 41, 37, 
                    35, 41, 39, 41, 42, 42, 36, 37, 37, 39,
                    42, 35, 36, 41, 41, 41, 39, 39, 35, 39
 ],
    'harness_size': [ 58, 58, 52, 58, 57, 52, 55, 53, 49, 54,
                59, 56, 53, 58, 57, 58, 56, 51, 50, 59,
                59, 59, 55, 50, 55, 52, 53, 54, 61, 56,
                55, 60, 57, 56, 61, 58, 53, 57, 57, 55,
                60, 51, 52, 56, 55, 57, 58, 57, 51, 59
                ]
}

# Conviértelo en una tabla usando pandas.
dataset = pandas.DataFrame(data)

# Imprime los datos 
# En python normal escribiríamos 
# print(dataset)
# pero en los cuadernos de Jupyter, si simplemente escribimos el nombre 
# de la variable y se imprime muy bien
dataset
```

Seleccione un modelo:

Utilizaremos el modelo OLS

```python
# Cargue una biblioteca para que haga el trabajo duro por nosotros
import statsmodels.formula.api as smf

# Primero, definimos nuestra fórmula usando una sintaxis especial
# Esto dice que boot_size se explica por harness_size
formula = "boot_size ~ harness_size"

# Cree el modelo, pero no lo entrene todavía
model = smf.ols(formula = formula, data = dataset)

# Tenga en cuenta que hemos creado nuestro modelo pero no 
# tiene parámetros internos establecidos todavía
if not hasattr(model, 'params'):
    print("Modelo seleccionado pero no tiene parámetros configurados. ¡Tenemos que entrenarlo!")
```


Entrenar a nuestro modelo:

Los modelos OLS tienen dos parámetros (una pendiente y una compensación), pero aún no se han establecido en nuestro modelo. Necesitamos entrenar ( ajustar ) nuestro modelo para encontrar estos valores para que el modelo pueda estimar de manera confiable el tamaño de las botas de los perros en función del tamaño de su arnés.

```python
# Cargue algunas bibliotecas para que hagan el trabajo duro por nosotros
import graphing 

# Entrena (ajusta) el modelo para que cree una línea que 
# se ajusta a nuestros datos. Este método hace el trabajo duro para 
# nosotros. Veremos cómo funciona este método en una unidad posterior.
fitted_model = model.fit()

# Imprimir información sobre nuestro modelo ahora que se ha ajustado
print("Se han encontrado los siguientes parámetros del modelo:\n" +
        f"Linea de pendiente: {fitted_model.params[1]}\n"+
        f"Linea de intersepto: {fitted_model.params[0]}")
```


Observe cómo el entrenamiento del modelo establece sus parámetros. Podríamos interpretarlos directamente, pero es más sencillo verlo como un gráfico:


```python
import graphing

# Mostrar un gráfico del resultado
# No te preocupes por cómo funciona esto por ahora
# graphing.scatter_2D(dataset,    label_x="harness_size", 
#                                 label_y="boot_size",
#                                 trendline=lambda x: fitted_model.params[1] * x + fitted_model.params[0]
#                                 )


import matplotlib.pyplot as plt
import numpy as np

m, b = np.polyfit(dataset["harness_size"], dataset["boot_size"], 1)

plt.scatter(dataset["harness_size"], dataset["boot_size"])
plt.plot(dataset["harness_size"], m*dataset["harness_size"] + b, color='red')
plt.xlabel("harness_size")
plt.ylabel("boot_size")
plt.show()
```

El gráfico anterior muestra nuestros datos originales como círculos atravesados por una línea roja. La línea roja muestra nuestro modelo .

Podemos mirar esta línea para entender nuestro modelo. Por ejemplo, podemos ver que a medida que aumenta el tamaño del arnés, también lo hará el tamaño estimado de la bota.


Usa el modelo:

Ahora que hemos terminado el entrenamiento, podemos usar nuestro modelo para predecir el tamaño de las botas de un perro a partir del tamaño de su arnés.

Por ejemplo, al mirar la línea roja, podemos ver que un tamaño de arnés de 52.5(eje x) corresponde a un tamaño de bota de aproximadamente 36.5(eje y).

```python
# harness_size indica el tamaño del arnés que nos interesa
harness_size = { 'harness_size' : [52.5] }

# Usa el modelo para predecir qué tamaño de botas calzará el perro
approximate_boot_size = fitted_model.predict(harness_size)

# Imprime el resultado
print("Estimado approximate_boot_size:")
print(approximate_boot_size[0])
```

Cargue el conjunto de datos:

Comencemos abriendo el conjunto de datos desde el archivo.

```python
import pandas as pd

# Cargue un archivo que contenga los tamaños de botas y arneses del perro
data = pd.read_csv('C:/Users/santi/OneDrive/Escritorio/Desktop/doggy-boot-harness.csv',delimiter=',',header='infer')

# Imprime las primeras filas
data.head()
```

    Crear y entrenar un modelo
Como hemos hecho antes, crearemos un modelo de regresión lineal simple y lo entrenaremos en nuestro conjunto de datos.

```python
import statsmodels.formula.api as smf

# Ajustar un modelo simple que encuentre una relación lineal
# entre el tamaño de la bota y el tamaño del arnés, que podemos usar más adelante 
# para predecir el tamaño de bota de un perro, dado el tamaño de su arnés

model = smf.ols(formula = "boot_size ~ harness_size", data = data).fit()

print("Modelo entrenado!")
```


    Guardar y cargar un modelo
Nuestro modelo está listo para usar, pero aún no lo necesitamos. Guardémoslo en el disco.

```python
import joblib

model_filename = './avalanche_dog_boot_model.pkl'
joblib.dump(model, model_filename)

print("Modelo guardado!")
```

    Cargar nuestro modelo es igual de fácil:

```python
model_loaded = joblib.load(model_filename)

print("Hemos cargado un modelo con los siguientes parámetros:")
print(model_loaded.params)
```

    Ponlo junto 
En nuestro sitio web, querremos tomar el arnés del perro de nuestro cliente y luego calcular el tamaño de la bota de su perro usando el modelo que ya hemos entrenado. Juntemos todo aquí para crear una función que cargue el modelo desde el disco y luego lo use para predecir la altura del tamaño de la bota del perro de nuestro cliente.

```python
# Escribamos una función que cargue y use nuestro modelo
def load_model_and_predict(harness_size):
    '''
    Esta función carga un modelo previamente entrenado. 
    Utiliza el modelo con el tamaño del arnés del perro del cliente 
    para predecir el tamaño de botas que se ajusten a ese perro.

    harness_size: El tamaño del arnés del perro, en cm.
    '''

    # Cargue el modelo desde el archivo e imprima información básica sobre él
    loaded_model = joblib.load(model_filename)

    print("Hemos cargado un modelo con los siguientes parámetros:")
    print(loaded_model.params)

    # Preparar datos para el modelo
    inputs = {"harness_size":[harness_size]} 

    # Use el modelo para hacer una prediccion
    predicted_boot_size = loaded_model.predict(inputs)[0]

    return predicted_boot_size

# Practique usando nuestro modelo
predicted_boot_size = load_model_and_predict(45)

print("Tamaño previsto de la bota del perro:", predicted_boot_size)
```

    Uso en el mundo real
Lo hicimos; podemos predecir el tamaño de la bota de un perro de avalanchas en función del tamaño de su arnés. Nuestro último paso es usar esto para advertir a las personas si podrían estar comprando botas para perros del tamaño incorrecto.

 Como ejemplo, haremos una función que acepte el tamaño del arnés, el tamaño de las botas seleccionadas y devuelva un mensaje para el cliente. Integraríamos esta función en nuestra tienda online.

```python
def check_size_of_boots(selected_harness_size, selected_boot_size):
    '''
    Calcula si el cliente ha elegido un par de botas para perros que son de un tamaño razonable. 
    Esto funciona estimando la bota real del perro. tamaño del tamaño de su arnés.

    Esto devuelve un mensaje para el cliente que debe mostrarse antes completan su pago 

    selected_harness_size: El tamaño del arnés que el cliente quiere comprar.
    selected_boot_size: El tamaño de las botas para perros que el cliente quiere comprar.
    '''

    # Calcule el tamaño de la bota del perro del cliente
    estimated_boot_size = load_model_and_predict(selected_harness_size)

    # Redondee al número entero más cercano porque no vendemos tamaños parciales
    estimated_boot_size = int(round(estimated_boot_size))

    # Comprobar si la talla de bota seleccionada es la adecuada
    if selected_boot_size == estimated_boot_size:
        # Las botas seleccionadas probablemente estén bien
        return f"¡Gran elección! Creemos que estas botas le quedarán bien a tu perro de avalancha.."

    if selected_boot_size < estimated_boot_size:
        # Las botas seleccionadas pueden ser demasiado pequeñas
        return "Las botas que ha seleccionado pueden ser DEMASIADO PEQUEÑAS para un perro, muy "\
               f"grande como el tuyo. Recomendamos un tamaño de botas para perros de {estimated_boot_size}."

    if selected_boot_size > estimated_boot_size:
        # Las botas seleccionadas pueden ser demasiado grandes
        return "Las botas que ha seleccionado pueden ser DEMASIADO GRANDES para un perro, muy "\
               f"pequeño como el tuyo. Recomendamos un tamaño de botas para perros de {estimated_boot_size}."
    

# Practique el uso de nuestro nuevo sistema de alerta
check_size_of_boots(selected_harness_size=50, selected_boot_size=33)
```
