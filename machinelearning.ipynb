Preparando datos:

Lo primero que hacemos con un modelo es cargar datos.

```python
import pandas as pd

# Hacer un diccionario de datos para tama√±os de botas
# y tama√±o del arn√©s en cm
data = {
    'boot_size' : [ 39, 38, 37, 39, 38, 35, 37, 36, 35, 40, 
                    40, 36, 38, 39, 42, 42, 36, 36, 35, 41, 
                    42, 38, 37, 35, 40, 36, 35, 39, 41, 37, 
                    35, 41, 39, 41, 42, 42, 36, 37, 37, 39,
                    42, 35, 36, 41, 41, 41, 39, 39, 35, 39
 ],
    'harness_size': [ 58, 58, 52, 58, 57, 52, 55, 53, 49, 54,
                59, 56, 53, 58, 57, 58, 56, 51, 50, 59,
                59, 59, 55, 50, 55, 52, 53, 54, 61, 56,
                55, 60, 57, 56, 61, 58, 53, 57, 57, 55,
                60, 51, 52, 56, 55, 57, 58, 57, 51, 59
                ]
}

# Convi√©rtelo en una tabla usando pandas.
dataset = pandas.DataFrame(data)

# Imprime los datos 
# En python normal escribir√≠amos 
# print(dataset)
# pero en los cuadernos de Jupyter, si simplemente escribimos el nombre 
# de la variable y se imprime muy bien
dataset
```

Seleccione un modelo:

Utilizaremos el modelo OLS

```python
# Cargue una biblioteca para que haga el trabajo duro por nosotros
import statsmodels.formula.api as smf

# Primero, definimos nuestra f√≥rmula usando una sintaxis especial
# Esto dice que boot_size se explica por harness_size
formula = "boot_size ~ harness_size"

# Cree el modelo, pero no lo entrene todav√≠a
model = smf.ols(formula = formula, data = dataset)

# Tenga en cuenta que hemos creado nuestro modelo pero no 
# tiene par√°metros internos establecidos todav√≠a
if not hasattr(model, 'params'):
    print("Modelo seleccionado pero no tiene par√°metros configurados. ¬°Tenemos que entrenarlo!")
```


Entrenar a nuestro modelo:

Los modelos OLS tienen dos par√°metros (una pendiente y una compensaci√≥n), pero a√∫n no se han establecido en nuestro modelo. Necesitamos entrenar ( ajustar ) nuestro modelo para encontrar estos valores para que el modelo pueda estimar de manera confiable el tama√±o de las botas de los perros en funci√≥n del tama√±o de su arn√©s.

```python
# Cargue algunas bibliotecas para que hagan el trabajo duro por nosotros
import graphing 

# Entrena (ajusta) el modelo para que cree una l√≠nea que 
# se ajusta a nuestros datos. Este m√©todo hace el trabajo duro para 
# nosotros. Veremos c√≥mo funciona este m√©todo en una unidad posterior.
fitted_model = model.fit()

# Imprimir informaci√≥n sobre nuestro modelo ahora que se ha ajustado
print("Se han encontrado los siguientes par√°metros del modelo:\n" +
        f"Linea de pendiente: {fitted_model.params[1]}\n"+
        f"Linea de intersepto: {fitted_model.params[0]}")
```


Observe c√≥mo el entrenamiento del modelo establece sus par√°metros. Podr√≠amos interpretarlos directamente, pero es m√°s sencillo verlo como un gr√°fico:


```python
import graphing

# Mostrar un gr√°fico del resultado
# No te preocupes por c√≥mo funciona esto por ahora
# graphing.scatter_2D(dataset,    label_x="harness_size", 
#                                 label_y="boot_size",
#                                 trendline=lambda x: fitted_model.params[1] * x + fitted_model.params[0]
#                                 )


import matplotlib.pyplot as plt
import numpy as np

m, b = np.polyfit(dataset["harness_size"], dataset["boot_size"], 1)

plt.scatter(dataset["harness_size"], dataset["boot_size"])
plt.plot(dataset["harness_size"], m*dataset["harness_size"] + b, color='red')
plt.xlabel("harness_size")
plt.ylabel("boot_size")
plt.show()
```

El gr√°fico anterior muestra nuestros datos originales como c√≠rculos atravesados por una l√≠nea roja. La l√≠nea roja muestra nuestro modelo .

Podemos mirar esta l√≠nea para entender nuestro modelo. Por ejemplo, podemos ver que a medida que aumenta el tama√±o del arn√©s, tambi√©n lo har√° el tama√±o estimado de la bota.


Usa el modelo:

Ahora que hemos terminado el entrenamiento, podemos usar nuestro modelo para predecir el tama√±o de las botas de un perro a partir del tama√±o de su arn√©s.

Por ejemplo, al mirar la l√≠nea roja, podemos ver que un tama√±o de arn√©s de 52.5(eje x) corresponde a un tama√±o de bota de aproximadamente 36.5(eje y).

```python
# harness_size indica el tama√±o del arn√©s que nos interesa
harness_size = { 'harness_size' : [52.5] }

# Usa el modelo para predecir qu√© tama√±o de botas calzar√° el perro
approximate_boot_size = fitted_model.predict(harness_size)

# Imprime el resultado
print("Estimado approximate_boot_size:")
print(approximate_boot_size[0])
```

Cargue el conjunto de datos:

Comencemos abriendo el conjunto de datos desde el archivo.

```python
import pandas as pd

# Cargue un archivo que contenga los tama√±os de botas y arneses del perro
data = pd.read_csv('C:/Users/santi/OneDrive/Escritorio/Desktop/doggy-boot-harness.csv',delimiter=',',header='infer')

# Imprime las primeras filas
data.head()
```

    Crear y entrenar un modelo
Como hemos hecho antes, crearemos un modelo de regresi√≥n lineal simple y lo entrenaremos en nuestro conjunto de datos.

```python
import statsmodels.formula.api as smf

# Ajustar un modelo simple que encuentre una relaci√≥n lineal
# entre el tama√±o de la bota y el tama√±o del arn√©s, que podemos usar m√°s adelante 
# para predecir el tama√±o de bota de un perro, dado el tama√±o de su arn√©s

model = smf.ols(formula = "boot_size ~ harness_size", data = data).fit()

print("Modelo entrenado!")
```


    Guardar y cargar un modelo
Nuestro modelo est√° listo para usar, pero a√∫n no lo necesitamos. Guard√©moslo en el disco.

```python
import joblib

model_filename = './avalanche_dog_boot_model.pkl'
joblib.dump(model, model_filename)

print("Modelo guardado!")
```

    Cargar nuestro modelo es igual de f√°cil:

```python
model_loaded = joblib.load(model_filename)

print("Hemos cargado un modelo con los siguientes par√°metros:")
print(model_loaded.params)
```

    Ponlo junto 
En nuestro sitio web, querremos tomar el arn√©s del perro de nuestro cliente y luego calcular el tama√±o de la bota de su perro usando el modelo que ya hemos entrenado. Juntemos todo aqu√≠ para crear una funci√≥n que cargue el modelo desde el disco y luego lo use para predecir la altura del tama√±o de la bota del perro de nuestro cliente.

```python
# Escribamos una funci√≥n que cargue y use nuestro modelo
def load_model_and_predict(harness_size):
    '''
    Esta funci√≥n carga un modelo previamente entrenado. 
    Utiliza el modelo con el tama√±o del arn√©s del perro del cliente 
    para predecir el tama√±o de botas que se ajusten a ese perro.

    harness_size: El tama√±o del arn√©s del perro, en cm.
    '''

    # Cargue el modelo desde el archivo e imprima informaci√≥n b√°sica sobre √©l
    loaded_model = joblib.load(model_filename)

    print("Hemos cargado un modelo con los siguientes par√°metros:")
    print(loaded_model.params)

    # Preparar datos para el modelo
    inputs = {"harness_size":[harness_size]} 

    # Use el modelo para hacer una prediccion
    predicted_boot_size = loaded_model.predict(inputs)[0]

    return predicted_boot_size

# Practique usando nuestro modelo
predicted_boot_size = load_model_and_predict(45)

print("Tama√±o previsto de la bota del perro:", predicted_boot_size)
```

    Uso en el mundo real
Lo hicimos; podemos predecir el tama√±o de la bota de un perro de avalanchas en funci√≥n del tama√±o de su arn√©s. Nuestro √∫ltimo paso es usar esto para advertir a las personas si podr√≠an estar comprando botas para perros del tama√±o incorrecto.

 Como ejemplo, haremos una funci√≥n que acepte el tama√±o del arn√©s, el tama√±o de las botas seleccionadas y devuelva un mensaje para el cliente. Integrar√≠amos esta funci√≥n en nuestra tienda online.

```python
def check_size_of_boots(selected_harness_size, selected_boot_size):
    '''
    Calcula si el cliente ha elegido un par de botas para perros que son de un tama√±o razonable. 
    Esto funciona estimando la bota real del perro. tama√±o del tama√±o de su arn√©s.

    Esto devuelve un mensaje para el cliente que debe mostrarse antes completan su pago 

    selected_harness_size: El tama√±o del arn√©s que el cliente quiere comprar.
    selected_boot_size: El tama√±o de las botas para perros que el cliente quiere comprar.
    '''

    # Calcule el tama√±o de la bota del perro del cliente
    estimated_boot_size = load_model_and_predict(selected_harness_size)

    # Redondee al n√∫mero entero m√°s cercano porque no vendemos tama√±os parciales
    estimated_boot_size = int(round(estimated_boot_size))

    # Comprobar si la talla de bota seleccionada es la adecuada
    if selected_boot_size == estimated_boot_size:
        # Las botas seleccionadas probablemente est√©n bien
        return f"¬°Gran elecci√≥n! Creemos que estas botas le quedar√°n bien a tu perro de avalancha.."

    if selected_boot_size < estimated_boot_size:
        # Las botas seleccionadas pueden ser demasiado peque√±as
        return "Las botas que ha seleccionado pueden ser DEMASIADO PEQUE√ëAS para un perro, muy "\
               f"grande como el tuyo. Recomendamos un tama√±o de botas para perros de {estimated_boot_size}."

    if selected_boot_size > estimated_boot_size:
        # Las botas seleccionadas pueden ser demasiado grandes
        return "Las botas que ha seleccionado pueden ser DEMASIADO GRANDES para un perro, muy "\
               f"peque√±o como el tuyo. Recomendamos un tama√±o de botas para perros de {estimated_boot_size}."
    

# Practique el uso de nuestro nuevo sistema de alerta
check_size_of_boots(selected_harness_size=50, selected_boot_size=33)
```
